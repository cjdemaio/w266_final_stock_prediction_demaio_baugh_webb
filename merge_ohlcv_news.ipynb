```python
import pandas as pd
```

```python
# load data
news_df = pd.read_csv('news_with_sentiment.csv')

ohlcv_df = pd.read_csv('sp100_ohlcv_2018_2023.csv')
```

```python
ohlcv_df = ohlcv_df[['Ticker', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
ohlcv_df.head()
```

```python
ohlcv_df["Date"] = pd.to_datetime(ohlcv_df["Date"], format='mixed')

# Truncate the time, keeping only the date part
ohlcv_df["Date"] = ohlcv_df["Date"].dt.date
```

```python
news_df.head()
```

```python
# Define a mapping for sentiment labels to numerical values
sentiment_mapping = {
    "negative": -1,
    "neutral": 0,
    "positive": 1
}

# Apply the mapping to the Sentiment_label column
news_df["Sentiment_numeric"] = news_df["Sentiment_label"].map(sentiment_mapping)
```

```python
news_df["Date"] = pd.to_datetime(news_df["Date"], format='mixed')

# Truncate the time, keeping only the date part
news_df["Date"] = news_df["Date"].dt.date
```

```python
news_df.head(20)
```

```python
# Group by 'Date' and 'Stock_symbol' and compute the average sentiment
average_sentiment = (
    news_df.groupby(["Date", "Stock_symbol"])["Sentiment_numeric"]
    .mean()
    .reset_index()
)

# Rename the resulting column for clarity
average_sentiment.rename(columns={"Sentiment_numeric": "Average_sentiment"}, inplace=True)

# Add the 'news_day' binary variable (1 if there was news, else 0)
average_sentiment["news_day"] = 1

# Create a full range of dates for each stock
all_dates = pd.date_range(news_df["Date"].min(), news_df["Date"].max())
all_stocks = news_df["Stock_symbol"].unique()
full_index = pd.MultiIndex.from_product([all_dates, all_stocks], names=["Date", "Stock_symbol"])

# Reindex the DataFrame to ensure all stock-date combinations are present
average_sentiment = average_sentiment.set_index(["Date", "Stock_symbol"]).reindex(full_index, fill_value=0).reset_index()

# Fill missing 'Average_sentiment' values with NaN (no news on that day)
average_sentiment.loc[average_sentiment["news_day"] == 0, "Average_sentiment"] = None
```

```python
average_sentiment[(average_sentiment['Stock_symbol'] == 'AAPL') & average_sentiment['news_day'] == 1]
```

```python
ohlcv_df.head()
```

```python
average_sentiment.rename(columns={"Stock_symbol": "Ticker"}, inplace=True)
```

```python
average_sentiment["Date"] = pd.to_datetime(average_sentiment["Date"])
ohlcv_df["Date"] = pd.to_datetime(ohlcv_df["Date"])
```

```python
# Perform the merge
merged_df = pd.merge(
    ohlcv_df,                 # The OHLCV DataFrame
    average_sentiment,        # The average sentiment DataFrame
    how="left",               # Use left join to keep all rows from ohlcv_df
    on=["Date", "Ticker"]     # Merge on Date and Ticker
)


# Display the merged DataFrame
merged_df.head()
```

```python
merged_df[(merged_df['Ticker'] == 'AAPL')]
```

```python
merged_df.to_csv('combined_sp100.csv')
```

```python
news_event_df = pd.read_csv('sp100_news_2018_2023_events_added.csv')
```

```python
news_event_df["Date"] = pd.to_datetime(news_event_df["Date"], format='mixed')

# Truncate the time, keeping only the date part
news_event_df["Date"] = news_event_df["Date"].dt.date
```

```python
news_event_df.rename(columns={"Stock_symbol": "Ticker"}, inplace=True)
news_event_df.head()
```

```python
label_map = {
    'I-A': 0, # Acquisition
    'I-CT': 1, # Clinical Trial
    'I-RD': 2, # Regular Dividend
    'I-DC': 3, # Dividend Cut
    'I-DI': 4, # Dividend Increase
    'I-GI': 5, # Guidance Increase
    'I-NC': 6, # New Contract
    'I-RSS': 7, # Reverse Stock Split
    'I-SD': 8, # Special Dividend
    'I-SR': 9, # Stock Repurchase
    'I-SS': 10, # Stock Split(SS)
    'O': 11, # No Event
    'I-GC': 12, #Unknown Label
}
```

```python
news_event_df['business_event'].value_counts()
```

```python
# Step 1: Create placeholder columns for one-hot encoding
for category in label_map.keys():
    news_event_df[category] = 0

# Step 2: Update the placeholder columns where business_event matches the code
for category, code in label_map.items():
    news_event_df.loc[news_event_df['business_event'] == code, category] = 1

# Step 3: Aggregate the one-hot columns by Date and Ticker using max()
# Select numeric columns for aggregation
one_hot_columns = list(label_map.keys())
aggregated = news_event_df.groupby(['Date', 'Ticker'], as_index=False)[one_hot_columns].max()

# Add back non-numeric columns
aggregated = pd.merge(news_event_df[['Date', 'Ticker']].drop_duplicates(), aggregated, on=['Date', 'Ticker'])
aggregated.head()
```

```python
# List of one-hot encoded indicator columns
indicator_columns = list(label_map.keys())

# Create a filter to check if more than one indicator variable is set to 1
filtered = aggregated[aggregated[indicator_columns].sum(axis=1) > 1]
filtered.shape
```
